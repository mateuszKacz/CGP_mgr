```python
x = [0.001*x for x in range(0,10000)]
```

```python
import matplotlib.pyplot as plt
from math import exp
```

```python
for k in range (1,10):
    for T in range(1,100,10):
        y = [1/k*exp(-val*k/T) for val in x]
        print(f'k: {k} \t T: {T}')
        plt.plot(x,y)
        plt.show()
        
```
